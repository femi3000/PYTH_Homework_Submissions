```
"""
basic operations II
"""

import numpy as np

arr = np.array([[1, 5, 6],
                [4, 7, 2],
                [3, 1, 9]])

# maximum element of array
print("Largest element is:", np.amax(arr))

# minimum element of array
print("Smallest element is:", np.amin(arr))

# maximum element per row
print("Row-wise maximum elements:", [np.amax(arr[0]), np.amax(arr[1]), np.amax(arr[2])])

# minimum element per col
print("Column-wise minimum elements:",
      [np.amin(arr[0]), np.amin(arr[1]), np.amin(arr[2])])

# sum of array elements
print("Sum of all array elements:", np.sum(arr))

# cumulative sum along each row
print("Cumulative sum along each row:\n", np.sum(arr[0]), np.sum(arr[1]), np.sum(arr[2]))

print("------------")

c = np.array([[1, 2],
              [3, 4]])
d = np.array([[4, 3],
              [2, 1]])

# add arrays
c1 = np.sum(c)
d1 = np.sum(d)
print("Array sum:\n", (c1 + d1))

print("------------")

# multiply arrays (elementwise multiplication)
c2 = (c*d)
#Not sure how to get the table from here 
print("Array multiplication:\n", None)
```
