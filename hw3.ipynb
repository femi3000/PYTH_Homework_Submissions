```
"""
Grades
"""
# Here's a dict containing students' grades from the semester's assignments.

students = {
  'Ashton': [86, 45, 98, 100],
  'Sierra': [100, 100, 100, 100],
  'Zach': [38, 49, 90, 87],
  'Manuel': [98, 92, 86, 100],
  'Felicia': [94, 87, 89, 95],
  'Ankur': [75, 77, 77, 85],
  'Ananya': [98, 94, 87, 92],
  'Nick': [79, 84, 89, 90],
  'Olivia': [83, 91, 69, 85],
  'Molly': [83, 74, 72, 90]
}

# p 1.1) Create a dict called num_grades to store each student's average numerical grade for the semester.

for k, v in students.items():
  avg = {k: (sum(v) / len(v))}
  students.update(avg)
num_grades = students
print(num_grades)
print("--------")


# p 1.2) Create a dict called letter_grades to store each student's final letter grade for the semester. Use A, B, C, D, and F as grades per the standard grading scale.

# letter_grades = 

for student, grade in num_grades.items():
  if 100 >= grade >= 90:
    num_grades.update({student: "A"})
  elif 90 >= grade >= 80:
    num_grades.update({student: "B"})
  elif 80 >= grade >= 70:
    num_grades.update({student: "C"})
  elif 70 >= grade >= 65:
    num_grades.update({student: "D"})
  else:
    num_grades.update({student: "F"})    

print(num_grades)

print("--------")

# p 1.3) Create a list of honor roll students (i.e. students who got A's).

# honor_roll 

honor_roll = [] 

for student, grade in num_grades.items():
  if grade is "A":
    honor_roll.append(student)
    
print(honor_roll)   
```
